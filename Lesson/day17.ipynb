# Working with Data (CSV and JSON) - Day 17

CSV and JSON are the two most common formats for storing and sharing data.Think of CSV like a spreadsheet table (rows and columns). Think of JSON like a nested filing system (dictionaries inside dictionaries).Knowing how to work with these formats is essential before moving to pandas.

## 1. What is CSV?CSV = Comma-Separated Values. It's just a text file where each line is a row and commas separate the columns.

## 2. Reading CSV FilesPython's `csv` module makes reading easy.

```python
import csv# Create a sample CSV file firstwith open("sales.csv", "w") as f:    f.write("item,quantity,price\n")    f.write("croissant,12,3.50\n")    f.write("muffin,8,2.00\n")    f.write("bagel,15,2.75\n")    f.write("cookie,23,1.50\n")# Now read it backwith open("sales.csv", "r") as f:    reader = csv.reader(f)    for row in reader:        print(row)# Output:# ['item', 'quantity', 'price']  (header row)# ['croissant', '12', '3.50']# ['muffin', '8', '2.00']# ['bagel', '15', '2.75']# ['cookie', '23', '1.50']
```

## 3. Reading CSV as DictionariesUsing `DictReader` gives you each row as a dictionary with the header row as keys.

```python
import csvwith open("sales.csv", "r") as f:    reader = csv.DictReader(f)    for row in reader:        print(row)# Output:# {'item': 'croissant', 'quantity': '12', 'price': '3.50'}# {'item': 'muffin', 'quantity': '8', 'price': '2.00'}# {'item': 'bagel', 'quantity': '15', 'price': '2.75'}# {'item': 'cookie', 'quantity': '23', 'price': '1.50'}
```

```python
# Now we can calculate totals easilyimport csvtotalRevenue = 0.0with open("sales.csv", "r") as f:    reader = csv.DictReader(f)    for row in reader:        qty = int(row['quantity'])        price = float(row['price'])        revenue = qty * price        totalRevenue += revenue        print(f"{row['item']}: ${revenue:.2f}")print(f"Total: ${totalRevenue:.2f}")
```

## 5. What is JSON?JSON = JavaScript Object Notation. It looks just like a Python dictionary (or a list of dicts).It's the most common format for web APIs and data exchange.

```python
import json# Writing JSONbakeryData = {    "name": "Sweet Treats Bakery",    "items": [        {"name": "croissant", "price": 3.50, "stock": 12},        {"name": "muffin", "price": 2.00, "stock": 8}    ],    "isOpen": True}with open("bakery.json", "w") as f:    json.dump(bakeryData, f, indent=2)print('Written to bakery.json')
```

```python
import json# Reading JSONwith open("bakery.json", "r") as f:    data = json.load(f)print(data['name'])           # Sweet Treats Bakeryprint(data['items'][0]['name'])  # croissantprint(data['isOpen'])         # True# json.loads() converts a JSON STRING to PythonjsonString = '{"name": "Test", "value": 42}'parsed = json.loads(jsonString)print(parsed['name'])  # Test# json.dumps() converts Python to a JSON STRINGpythonDict = {"name": "Gaby", "score": 95}jsonStr = json.dumps(pythonDict)print(jsonStr)  # {"name": "Gaby", "score": 95}
```

## Summary- **CSV**: Tabular data, like a spreadsheet.- `csv.reader()` / `csv.DictReader()` - read CSV files.- `csv.writer()` / `csv.DictWriter()` - write CSV files.- **JSON**: Hierarchical data, like nested dicts.- `json.load()` / `json.loads()` - read JSON.- `json.dump()` / `json.dumps()` - write JSON.- **Pipeline**: Read -> Process -> Write is a common pattern.

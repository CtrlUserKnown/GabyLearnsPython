## 2. Creating Specific DatesYou can create any date you want.

```python
from datetime import date# The bakery opened on March 15, 2024bakeryOpening = date(2024, 3, 15)print(bakeryOpening)  # 2024-03-15# A specific time (hours, minutes, seconds)from datetime import timeopeningTime = time(7, 0, 0)  # 7:00 AMclosingTime = time(19, 0, 0)  # 7:00 PMprint(f"Open: {openingTime} to {closingTime}")
```

## 3. Formatting Dates (for humans)Use `strftime()` to turn dates into readable strings.Think of it like a photo filter. The date is the photo, and the format string is the filter that changes how it looks.

```python
from datetime import datetimenow = datetime.now()print(now.strftime("%Y-%m-%d"))           # 2026-06-04print(now.strftime("%B %d, %Y"))          # June 04, 2026print(now.strftime("%A, %I:%M %p"))       # Thursday, 02:30 PMprint(now.strftime("%m/%d/%y"))           # 06/04/26# Common format codes:# %Y - 4-digit year    %y - 2-digit year# %m - month (01-12)   %B - full month name# %d - day (01-31)     %A - full weekday name# %H - 24-hour hour    %I - 12-hour hour# %M - minute          %p - AM/PM
```

## 4. Parsing Dates (from strings)`strptime()` turns a string into a date object. It's the reverse of `strftime()`.

```python
from datetime import datetime# Convert a date string into a datetime objectdateStr = "2026-06-04"parsed = datetime.strptime(dateStr, "%Y-%m-%d")print(parsed)            # 2026-06-04 00:00:00print(parsed.weekday())  # 3 (Monday=0, Sunday=6, so Thursday=3)# More complex parsingdateStr2 = "June 4, 2026 at 2:30 PM"parsed2 = datetime.strptime(dateStr2, "%B %d, %Y at %I:%M %p")print(parsed2)  # 2026-06-04 14:30:00
```

## 6. Working with Time Zones (Brief Intro)Time zones can be tricky. Python's basic `datetime` is "naive" (no timezone). For serious work, use the `pytz` library or `zoneinfo` (Python 3.9+).

```python
# Python 3.9+ has built-in zoneinfotry:    from zoneinfo import ZoneInfo    from datetime import datetime    ny_time = datetime.now(ZoneInfo("America/New_York"))    london_time = datetime.now(ZoneInfo("Europe/London"))    print(f"New York: {ny_time.strftime('%I:%M %p')}")    print(f"London:   {london_time.strftime('%I:%M %p')}")except ImportError:    print("zoneinfo not available (use Python 3.9+)")
```

# Python Data Types - Day 2

Python uses different data types to know what kind of information you are working with.
Think of these as different types of "boxes" for your data.

## 1. Strings
For text like words, sentences, or even just one letter. Always put these inside quotes.
These are like labels on a folder — words that tell you what something is.

```
x = "Hello World"
print(type(x))
```

## 2. Ints (Integers)
For whole numbers that don't have decimals. Good for counting things.
These are like Whole Apples. You can't have 1.5 whole apples, only 1, 2, or 3.

```
x = 20
print(type(x))
```

## 3. Floats (Floating Point)
For numbers that DO have decimals. Useful for measurements or prices.
These are more like Measuring Cups. You can have 1.5 cups or 0.75 cups.

```
x = 20.5
print(type(x))
```

## 4. Booleans
For things that are only True or False. Like a light switch: it is either on or off.

```
isItRaining = True
print(type(isItRaining))
```

## 5. Lists
For when you want to store a bunch of things in one place.
Use [square brackets] and commas to separate them.
It's like a grocery bag where you put many items in one place.

```
shoppingList = ["apples", "milk", "eggs"]
print(shoppingList)
```

## Function Structure Overview
Functions are like "recipes". You define them once, and then you can use them whenever you want.

- `def` — tells Python you are starting a function
- `functionName` — the name you give it
- `(parameters)` — the things the function needs to do its job
- `:` — starts the body of the function
- indentation — everything inside MUST be tabbed/spaced over
- `return` — how the function gives back the result

```
def illustrateDataTypes(userName, age, currentPoints):
    welcome = "Hello " + userName
    print(welcome)

    yearsUntilOneHundred = 100 - age

    bonusMultiplier = 1.5
    totalScore = currentPoints * bonusMultiplier
    print("Your total score is: " + str(totalScore))

    isHighScore = totalScore > 100
    print("Is this a high score? " + str(isHighScore))

    summary = [userName, age, totalScore, isHighScore]
    return summary

print(illustrateDataTypes("Alice", 25, 80))
```

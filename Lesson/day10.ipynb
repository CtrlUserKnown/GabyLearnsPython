# Python Modules and Packages - Day 10

As your code grows, keeping everything in one file gets messy.
Modules and packages are Python's way of organizing code into separate files.
Think of it like a toolbox. Instead of dumping every tool into one giant drawer, you sort them into smaller boxes. The "wrench box" has all the wrenches, the "screwdriver box" has all the screwdrivers. When you need a wrench, you grab just that box.

## 1. What is a Module?
A module is just a `.py` file. Any Python file can be imported into another file.
When you `import` it, you get access to all the functions, variables, and classes inside it.

## 2. Importing a Module
There are a few ways to import:

```python
# Way 1: Import the whole module
import math

print(math.sqrt(16))   # 4.0
print(math.pi)         # 3.14159...
```

```python
# Way 2: Import specific things (saves typing)
from math import sqrt, pi

print(sqrt(25))   # 5.0  (no need for math. prefix)
print(pi)         # 3.14159...
```

```python
# Way 3: Give it a nickname (alias)
import math as m

print(m.sqrt(9))   # 3.0
```

## 3. The `if __name__ == "__main__"` Pattern
This is a very common pattern in Python. It checks if the file is being run directly or being imported.
Think of it like a stage door. When you run a file directly, it's the "main show" — the code runs. When you import it, it's like a backstage pass — you can use the tools, but the show doesn't start.

```python
# Save this as bakery_tools.py and run it

def calculateBakingTime(items, minutesPerItem):
    return items * minutesPerItem

def convertOvenTemp(celsius):
    return celsius * 9/5 + 32

if __name__ == "__main__":
    # This only runs when the file is executed directly
    print("Testing bakery tools...")
    print(calculateBakingTime(12, 15))  # 180
    print(convertOvenTemp(180))          # 356
```

## 4. What is a Package?
A package is a folder that contains multiple modules. The folder needs a special file called `__init__.py` to tell Python "this folder is a package."
Think of a package like a filing cabinet. Each drawer (module) contains related papers (code), and the `__init__.py` is like the label on the cabinet that says "open me."

## 5. Installing Third-Party Packages with pip
Python comes with a tool called `pip` that downloads packages from the internet.
Think of pip like a delivery service. You tell it what you want ("I need pandas"), and it brings it to your computer.

```python
# Run this in your terminal, not in Python:
# pip install pandas
# pip install requests
# pip install numpy
```

## 7. Python's Standard Library
Python comes with a huge collection of built-in modules called the "standard library." You don't need to install anything extra.
Some useful ones:
- `math` — advanced math functions
- `random` — generate random numbers
- `datetime` — work with dates and times
- `json` — read and write JSON data
- `csv` — read and write CSV files
- `os` — interact with your operating system
- `sys` — access system settings

## Summary
- A **module** is a `.py` file. A **package** is a folder of modules.
- `import module` — import the whole module.
- `from module import thing` — import just what you need.
- `import module as alias` — give it a shorter name.
- `if __name__ == "__main__"` — code that only runs when executed directly.
- `pip` installs third-party packages.
- Python's **standard library** has many useful modules built in.

# Building a Data Pipeline - Day 19

A data pipeline is a series of steps that takes raw data and transforms it into something useful.Think of it like a bakery assembly line. Flour goes in one end, bread comes out the other. Each station does one job.Today we'll build a complete pipeline from scratch.

## 1. The Pipeline PatternEvery pipeline follows the same structure:1. **Extract** - Get the data (read a file, fetch from API, etc.)2. **Transform** - Clean and process the data3. **Analyze** - Calculate insights4. **Output** - Save or display the resultsEach step should be its OWN function. This is called **modularity**.

```python
# Good code practice: Each function does ONE thing# Function names describe what they doimport csvimport jsonfrom collections import Counterdef extractSales(filepath):    """Read sales data from a CSV file."""    sales = []    with open(filepath, 'r') as f:        reader = csv.DictReader(f)        for row in reader:            sales.append(row)    return salesdef cleanSales(sales):    """Convert string values to numbers and validate data."""    cleaned = []    for row in sales:        try:            row['quantity'] = int(row['quantity'])            row['price'] = float(row['price'])            row['revenue'] = round(row['quantity'] * row['price'], 2)            cleaned.append(row)        except (ValueError, KeyError) as e:            print(f'Skipping bad row: {row} - Error: {e}')    return cleaned
```

```python
def analyzeSales(cleanedSales):    """Calculate summary statistics from sales data."""    if not cleanedSales:        return {}    totalRevenue = sum(row['revenue'] for row in cleanedSales)    totalItems = sum(row['quantity'] for row in cleanedSales)    avgPrice = totalRevenue / totalItems if totalItems > 0 else 0    bestSeller = max(cleanedSales, key=lambda r: r['quantity'])    return {        'totalRevenue': totalRevenue,        'totalItemsSold': totalItems,        'averagePrice': round(avgPrice, 2),        'bestSeller': bestSeller['item'],        'totalOrders': len(cleanedSales)    }def outputReport(analysis, sales, filepath):    """Save the analysis as a JSON report."""    report = {        'summary': analysis,        'details': sales,    }    with open(filepath, 'w') as f:        json.dump(report, f, indent=2)    return report
```

```python
# Now let's create some sample data and run our pipelinewith open("pipeline_sales.csv", "w") as f:    f.write("item,quantity,price\n")    f.write("croissant,12,3.50\n")    f.write("muffin,8,2.00\n")    f.write("bagel,15,2.75\n")    f.write("cookie,23,1.50\n")    f.write("bad data,not_a_number,3.00\n")# Run the full pipeline:rawData = extractSales('pipeline_sales.csv')cleanData = cleanSales(rawData)analysis = analyzeSales(cleanData)report = outputReport(analysis, cleanData, 'pipeline_report.json')print(json.dumps(report, indent=2))
```

## 4. Error Handling in a PipelinePipelines fail. Good pipelines handle failures gracefully.Each function should validate its inputs and handle errors without crashing the whole process.

```python
def runPipeline(inputFile, outputFile):    """Run the full pipeline with error handling."""    try:        print(f'Reading data from {inputFile}...')        rawData = extractSales(inputFile)        print('Cleaning data...')        cleanData = cleanSales(rawData)        if not cleanData:            print('ERROR: No valid data to analyze!')            return None        print('Analyzing...')        analysis = analyzeSales(cleanData)        print(f'Writing report to {outputFile}...')        report = outputReport(analysis, cleanData, outputFile)        print('Pipeline complete!')        return report    except FileNotFoundError:        print(f'ERROR: Could not find file {inputFile}')    except PermissionError:        print(f'ERROR: Permission denied for {inputFile}')    except Exception as e:        print(f'ERROR: Unexpected error: {e}')    return None# Run itresult = runPipeline('pipeline_sales.csv', 'pipeline_report.json')
```

## 6. Python Best Practices ChecklistWhen writing any code, ask yourself:1. **Does each function do ONE thing?** If it does two, split it.2. **Are my variable names clear?** `calculateTotal()` not `calc()`.3. **Do I have docstrings?** Every function explains what it does.4. **Do I handle errors?** What happens if the file doesn't exist?5. **Is my code consistent?** Same naming style everywhere.6. **No hardcoded values?** Use constants or parameters instead.7. **No commented-out code?** Delete it instead of leaving it as a tombstone.8. **Is it tested?** At least mentally trace through what it does.

## Summary- **Pipeline pattern**: Extract -> Transform -> Analyze -> Output.- **Each function does ONE thing** (modularity).- **Error handling**: Use try/except, validate inputs, fail gracefully.- **Testing**: Verify each piece independently.- **Best practices**: Clear names, docstrings, no hardcoded values, no dead code.

## 1. The `math` ModuleThe `math` module gives you access to advanced math functions.

```python
import mathprint(math.ceil(4.2))    # 5   (round up)print(math.floor(4.8))   # 4   (round down)print(math.round(4.5))   # 4   (banker's rounding, built-in)print(math.sqrt(144))    # 12.0print(math.pow(2, 10))   # 1024.0  (2 to the power of 10)print(math.pi)           # 3.141592653589793print(math.inf)          # inf (infinity)
```

## 3. The `statistics` ModuleBasic statistical functions built right in.

```python
import statisticsdailySales = [34, 28, 45, 32, 41, 29, 38]  # Sales from last weekprint(f"Mean (average):   {statistics.mean(dailySales):.1f}")print(f"Median (middle):  {statistics.median(dailySales)}")print(f"Mode (most common): {statistics.mode(dailySales)}")print(f"Stdev (spread):   {statistics.stdev(dailySales):.1f}")
```

## 4. Rounding and PrecisionBe careful with floats! They can behave unexpectedly due to how computers store numbers.

```python
# The float problemprint(0.1 + 0.2)  # 0.30000000000000004 (not 0.3!)# Solution: round() or use the decimal moduleprint(round(0.1 + 0.2, 2))  # 0.3from decimal import Decimalprint(Decimal("0.1") + Decimal("0.2"))  # 0.3 (exact!)
```

## 5. List of Useful Math FunctionsHere are some more handy tools from the `math` module.

```python
import mathprint(math.fabs(-10))      # 10.0 (absolute value as float)print(math.factorial(5))  # 120print(math.gcd(12, 8))     # 4 (greatest common divisor)print(math.isfinite(5))   # True (is it a normal number?)print(math.isnan(5))      # False (is it Not-a-Number?)# min() and max() are built-in (no math module needed)print(min(3, 7, 2, 9))    # 2print(max(3, 7, 2, 9))    # 9print(sum([1, 2, 3, 4]))  # 10
```

## 1. The `get()` Method (Review + Deep Dive)
`get()` is safer than using brackets `[]` because it won't crash if the key is missing.

```python
prices = {"croissant": 3.50, "muffin": 2.50}

# Safe way — returns None if missing
print(prices.get("croissant"))   # 3.5
print(prices.get("bagel"))       # None
print(prices.get("bagel", 0))    # 0 (default value if missing)
```

## 2. The `setdefault()` Method
`setdefault()` is like `get()`, but it also SETS the default value if the key is missing.
Think of it like a coat check. If someone doesn't have a ticket, you give them one and record it at the same time.

```python
sales = {}

# Each day, add sales for items
itemsSold = ["croissant", "muffin", "croissant", "bagel", "croissant", "muffin"]

for item in itemsSold:
    sales.setdefault(item, 0)  # Start at 0 if not seen before
    sales[item] += 1

print(sales)  # {'croissant': 3, 'muffin': 2, 'bagel': 1}
```

## 3. The `defaultdict` from collections
Python has a special dictionary called `defaultdict` that does the `setdefault` trick automatically.
Think of it like a vending machine that always has a default item ready if the slot is empty.

```python
from collections import defaultdict

# defaultdict with int: missing keys automatically get 0
sales = defaultdict(int)

itemsSold = ["croissant", "muffin", "croissant", "bagel", "croissant", "muffin"]

for item in itemsSold:
    sales[item] += 1  # No need for setdefault — it's automatic!

print(dict(sales))  # {'croissant': 3, 'muffin': 2, 'bagel': 1}
```

```python
# defaultdict with list: missing keys automatically get an empty list
from collections import defaultdict

ordersByCustomer = defaultdict(list)

ordersByCustomer["Alice"].append("croissant")
ordersByCustomer["Bob"].append("muffin")
ordersByCustomer["Alice"].append("coffee")  # Alice already exists, this appends

print(dict(ordersByCustomer))
# {'Alice': ['croissant', 'coffee'], 'Bob': ['muffin']}
```

## 4. The `Counter` from collections
Counter is a special dictionary for counting things. It's perfect when you need to know "how many of each?"
Think of it like a tally mark sheet.

```python
from collections import Counter

itemsSold = ["croissant", "muffin", "croissant", "bagel", "croissant", "muffin"]
counts = Counter(itemsSold)

print(counts)             # Counter({'croissant': 3, 'muffin': 2, 'bagel': 1})
print(counts["croissant"]) # 3
print(counts.most_common(2))  # [('croissant', 3), ('muffin', 2)]
```

## 5. Merging Dictionaries
There are several ways to combine dictionaries.

```python
# Method 1: update() — adds keys from one dict to another
morningSales = {"croissant": 10, "muffin": 5}
afternoonSales = {"bagel": 8, "cookie": 12}

morningSales.update(afternoonSales)
print(morningSales)  # {'croissant': 10, 'muffin': 5, 'bagel': 8, 'cookie': 12}
```

```python
# Method 2: | operator (Python 3.9+)
morningSales = {"croissant": 10, "muffin": 5}
afternoonSales = {"bagel": 8, "cookie": 12}

allSales = morningSales | afternoonSales
print(allSales)  # {'croissant': 10, 'muffin': 5, 'bagel': 8, 'cookie': 12}
```

## 6. Set Operations
Sets support math operations like union, intersection, and difference.

```python
morningItems = {"croissant", "muffin", "bagel"}
afternoonItems = {"muffin", "cookie", "scone"}

# Union — everything that was sold today
allItems = morningItems | afternoonItems
print("All:", allItems)  # {'croissant', 'muffin', 'bagel', 'cookie', 'scone'}

# Intersection — items sold both times
both = morningItems & afternoonItems
print("Both:", both)     # {'muffin'}

# Difference — items only sold in the morning
onlyMorning = morningItems - afternoonItems
print("Only morning:", onlyMorning)  # {'croissant', 'bagel'}
```

```python
# Checking if one set is a subset of another
inventory = {"flour", "sugar", "butter", "eggs", "vanilla"}
recipe = {"flour", "sugar", "eggs"}

print(recipe.issubset(inventory))  # True — we have everything for the recipe
print(inventory.issuperset(recipe))  # True — same check, other direction
```

## Summary
- `dict.get(key, default)` — safe access without crashing.
- `dict.setdefault(key, default)` — access and set default in one step.
- `defaultdict` — auto-defaults missing keys (great for counting and grouping).
- `Counter` — special dict built for counting.
- Merge dicts with `update()` or `|` operator.
- Set operations: `|` (union), `&` (intersection), `-` (difference), `issubset()`.

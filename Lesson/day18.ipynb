# Thinking in Arrays - Day 18

Before you learn pandas, you need to start thinking in terms of columns and arrays instead of individual values.Think of it like the difference between making cookies one at a time (a loop) and using a giant sheet pan that bakes them all at once (vectorized).This mental shift is the most important step before learning pandas.

## 1. Lists as Columns of DataInstead of storing one value at a time, store all values in a list and operate on the whole list.This is how pandas thinks.

```python
# Instead of individual variables...item1_price = 3.50item2_price = 2.00item3_price = 2.75# Think in columns:prices = [3.50, 2.00, 2.75, 1.50]quantities = [12, 8, 15, 23]names = ['croissant', 'muffin', 'bagel', 'cookie']# Now you can work with ALL the data at onceprint(f"Prices: {prices}")print(f"Quantities: {quantities}")
```

## 2. The Vectorized MindsetInstead of looping, think: "What if I could apply an operation to ALL items at once?"This is called **vectorized** thinking.

```python
# Loop way (one at a time)revenues_loop = []for i in range(len(prices)):    revenues_loop.append(prices[i] * quantities[i])# Vectorized way (all at once using list comprehension)revenues = [p * q for p, q in zip(prices, quantities)]print(revenues)  # [42.0, 16.0, 41.25, 34.5]
```

```python
# Zip is your friend - it pairs up elements from multiple listsnames = ['croissant', 'muffin', 'bagel', 'cookie']prices = [3.50, 2.00, 2.75, 1.50]for name, price in zip(names, prices):    print(f"{name}: ${price:.2f}")# Zip creates pairs: ('croissant', 3.50), ('muffin', 2.00), ...
```

## 3. Filtering with the Vectorized MindsetInstead of "check each item," think "which items meet this condition?"

```python
# Loop way:cheapItems_loop = []for name, price in zip(names, prices):    if price < 3.00:        cheapItems_loop.append(name)# Vectorized way (using list comprehension):cheapItems = [name for name, price in zip(names, prices) if price < 3.00]print(cheapItems)  # ['muffin', 'bagel', 'cookie']
```

## 4. Multiple Conditions on ArraysYou can combine conditions just like pandas does.Think of it like a spotlight. Each condition is a filter that only lets certain rows through.

```python
# Find items that sold more than 10 AND cost over $2.00quantities = [12, 8, 15, 23]prices = [3.50, 2.00, 2.75, 1.50]names = ['croissant', 'muffin', 'bagel', 'cookie']result = [    name    for name, qty, price in zip(names, quantities, prices)    if qty > 10 and price > 2.00]print(result)  # ['croissant', 'bagel']
```

## 5. The Problem with Plain Python ListsPlain Python lists are slow for large datasets. This is why libraries like NumPy and pandas exist.Think of it like counting grains of rice one by one (a Python list) vs. using a measuring cup (NumPy/pandas).

```python
# Plain Python: each operation creates a new listnumbers = list(range(1000000))doubled = [n * 2 for n in numbers]  # Creates a new listprint(f"First 5: {doubled[:5]}")  # [0, 2, 4, 6, 8]# NumPy does this MUCH faster (we'll learn this next):# import numpy as np# arr = np.array(range(1000000))# doubled = arr * 2  # No loop needed!
```

## 6. Thinking in Tables (The Pandas Mindset)A table (or DataFrame) is just a collection of columns that are all the same length.Each column is a list/array. The whole table is the collection of columns.

```python
# This is what a DataFrame looks like in your head:table = {    'item':    ['croissant', 'muffin', 'bagel', 'cookie'],    'price':   [3.50,       2.00,     2.75,    1.50],    'sold':    [12,         8,        15,       23],    'revenue': [42.0,       16.0,     41.25,    34.5],}# Later, pandas will let you do:# df = pandas.DataFrame(table)# df[df['sold'] > 10]  # Filter rows# df['revenue'].mean() # Average revenue
```

## Summary- **Think in columns**: Store data as lists (columns), not individual variables.- **Vectorized mindset**: Operate on all items at once instead of looping.- **`zip()`**: Pair up elements from multiple lists.- **Comprehensions**: Filter and transform data in one step.- **Lists are slow** for big data. NumPy and pandas solve this.- **DataFrame**: A collection of equal-length columns (a table).

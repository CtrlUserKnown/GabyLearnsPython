You already know the basics of functions. Today we unlock their full power.Think of functions like a Swiss Army knife. The basic version just cuts. But there are extra tools hidden inside — and once you know they exist, you wonder how you lived without them.

## 1. `*args` — Accept Any Number of ArgumentsThe `*args` parameter lets a function accept as many arguments as you throw at it.Think of it like a bakery box. Instead of saying "I want exactly 3 cookies," you can say "give me some cookies" and the box handles any number.

```python
def makePancakes(*toppings):    """Accept any number of toppings for pancakes."""    print("Making pancakes with:")    for topping in toppings:        print(f"  - {topping}")makePancakes("syrup", "butter")print()makePancakes("syrup", "butter", "berries", "whipped cream")
```

## 2. `**kwargs` — Accept Any Number of Keyword ArgumentsThe `**kwargs` parameter accepts any number of named arguments.Think of it like a custom order form. Instead of a fixed form, the customer can write whatever customizations they want.

```python
def buildOrder(**items):    """Build a custom order from keyword arguments."""    print("Order summary:")    for item, quantity in items.items():        print(f"  {item}: {quantity}")buildOrder(croissant=3, muffin=2, coffee=1)print()buildOrder(bagel=5, creamCheese=2)
```

## 3. Lambda Functions (Anonymous Functions)A lambda is a tiny function with no name — a "throwaway" function.Think of it like a sticky note. Instead of writing a full document (a `def` function), you jot down a quick note and throw it away after using it.

```python
# Regular functiondef double(x):    return x * 2# Same thing as a lambdadouble = lambda x: x * 2print(double(5))  # 10
```

```python
# Lambdas are most useful with sort(), filter(), and map()# Sort items by price (using a lambda to extract the price)menu = [("croissant", 3.50), ("muffin", 2.00), ("bagel", 2.75)]menu.sort(key=lambda item: item[1])print(menu)  # [('muffin', 2.0), ('bagel', 2.75), ('croissant', 3.5)]
```

## 4. The `map()` Function`map()` applies a function to every item in a list.Think of it like a conveyor belt. Each item goes in, gets processed, and comes out transformed.

```python
# Apply a 10% discount to all pricesprices = [3.50, 2.00, 2.75, 1.50]discounted = list(map(lambda p: round(p * 0.9, 2), prices))print(discounted)  # [3.15, 1.8, 2.48, 1.35]
```

## 5. The `filter()` Function`filter()` keeps only items where the function returns True.Think of it like a sieve. You pour everything in, and only the stuff that fits through stays.

```python
# Keep only items priced under $3.00menu = [("croissant", 3.50), ("muffin", 2.00), ("bagel", 2.75), ("cookie", 1.50)]cheapItems = list(filter(lambda item: item[1] < 3.00, menu))print(cheapItems)  # [('muffin', 2.0), ('bagel', 2.75), ('cookie', 1.5)]
```

## 6. Docstrings — Documenting Your FunctionsA docstring is a string right after a function definition that explains what it does.Think of it like an instruction manual. It's not the code itself, but the explanation of how to USE the code.

```python
def calculateBakingTime(items, minutesPerItem):    """Calculate total baking time for a batch of items.    Args:        items: Number of items to bake (int)        minutesPerItem: Minutes per item (float)    Returns:        Total baking time in minutes (float)    """    return items * minutesPerItemhelp(calculateBakingTime)  # Prints the docstring
```

## 7. Type Hints (Annotations)You can hint at what type a parameter should be. Python doesn't enforce it, but it helps readers (and editors) understand your code.Think of it like a label on a bin. It says "screws" so you know what goes in there, but nothing stops you from putting nails in.

```python
def calculateBakingTime(items: int, minutesPerItem: float) -> float:    """Calculate total baking time."""    return items * minutesPerItem# The hints tell you: items should be an int, minutesPerItem a float, returns a floatprint(calculateBakingTime(12, 15.0))  # 180.0
```

## Summary- `*args` — accept any number of positional arguments.- `**kwargs` — accept any number of keyword arguments.- `lambda` — create a small one-time function.- `map()` — apply a function to every item in a list.- `filter()` — keep only items that pass a test.- **Docstrings** — document what your function does.- **Type hints** — indicate what types a function expects.

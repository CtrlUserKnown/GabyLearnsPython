Text is everywhere in programming. Python has incredible tools for working with strings.Think of text like dough. You can shape it, cut it, combine it, and decorate it however you want.

## 1. f-Strings (The Best Way to Format Text)f-strings let you embed variables directly inside a string using `{}`.Think of it like a fill-in-the-blank worksheet. You write the sentence with blanks, and Python fills them in.

```python
itemName = "croissant"quantity = 12price = 3.50# Old way (concatenation)print("Sold " + str(quantity) + " " + itemName + "s for $" + str(price * quantity))# f-string way (much cleaner)print(f"Sold {quantity} {itemName}s for ${price * quantity}")
```

```python
# Formatting numbers inside f-stringsprice = 3.5print(f"Price: ${price:.2f}")       # Price: $3.50  (2 decimal places)score = 0.8765print(f"Score: {score:.1%}")       # Score: 87.7% (percentage)number = 1234567print(f"{number:,.0f}")            # 1,234,567  (comma separator)
```

## 2. The `join()` Method`join()` combines a list of strings into one string with a separator.Think of it like threading beads onto a string. The beads are your words, and the string is whatever separator you choose.

```python
items = ["croissant", "muffin", "bagel"]# Join with comma and spacemenu = ", ".join(items)print(menu)  # croissant, muffin, bagel# Join with newlinesreceipt = "\n".join(items)print(receipt)
```

## 3. The `split()` Method (Review + Deep Dive)`split()` breaks a string into a list. It's the opposite of `join()`.

```python
data = "croissant, muffin, bagel, cookie"items = data.split(", ")print(items)  # ['croissant', 'muffin', 'bagel', 'cookie']# Split on whitespace (default)sentence = "The bakery opens at 7 AM"words = sentence.split()print(words)  # ['The', 'bakery', 'opens', 'at', '7', 'AM']
```

## 4. The `strip()` Family`strip()` removes whitespace from both ends. `lstrip()` and `rstrip()` remove from left or right only.

```python
userInput = "   croissant   "print(repr(userInput))       # '   croissant   'print(repr(userInput.strip()))  # 'croissant'# Also works with specific charactersphone = "---555-1234---"print(phone.strip("-"))  # 555-1234
```

## 5. Checking What's in a StringPython has several built-in string checkers.

```python
# String check methodsprint("hello".isalpha())   # True  (all letters)print("123".isdigit())     # True  (all digits)print("abc123".isalnum())  # True  (letters and digits only)print("   ".isspace())     # True  (only whitespace)print("hello123".isalnum())  # Trueprint("hello!".isalnum())    # False (because of the !)
```

## 6. Finding and ReplacingUseful for cleaning and searching text.

```python
text = "The best bagel in the bakery is the blueberry bagel."# find() returns the index where it starts, or -1 if not foundprint(text.find("bagel"))       # 15print(text.find("pizza"))       # -1# replace() substitutes all occurrencesprint(text.replace("bagel", "muffin"))# "The best muffin in the bakery is the blueberry muffin."
```

## Summary- **f-strings**: `f"value is {x:.2f}"` — clean string formatting.- **`join()`**: Combine a list into a string.- **`split()`**: Break a string into a list.- **`strip()`**: Remove whitespace (or other chars) from ends.- **Check methods**: `isalpha()`, `isdigit()`, `isalnum()`, etc.- **`find()`**: Search for text. **`replace()`**: Replace text.- **Regex (`re` module)**: Pattern matching for advanced text search.
